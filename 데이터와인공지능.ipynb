데이터는 긴 숫자로 표현해야 한다.   
데이터는 숫자열이다.  
사람들은 ai가 매우 복잡할 것이라고 생각하는데 실제로는 직사각형 상태의 숫자가 있는 간단한 형태이다.  
입력값과 ai를 곱해준다.  곱할 때는 세로줄을 기준으로 한줄씩 곱한다. 글고 다 더해서 입력 값의 첫번째 자리에 넣는다. 다음에는 두번쨰 줄을 곱하고 더해  출력값의 두번쨰 자리에 넣는 과정을 반복하면 된다.  
입출력 값을 우리는 벡터라고 부른다.(입려 벡터, 출력 벡터) 가운데 ai는 행렬이라고 부른다.  
행렬 = 직사각형 벡터 = 숫자열(한줄짜리 행열)  
ai행렬이 하나만 있는 것이 아니라 여러개 있을 수 있다. 여러 개 있으면 같은 과정을 계속해서 반복하면 된다. 하나 있을 때부터 입력 벡터로부터 출ㄹ력 벡터까지의 거리가 딥해졌다. 우리는 이것을 딥러닝이라고 부른다. 초창기 때는 이것을 인공 신경망 즉, artificial neutral network (ann)이라고 불렀다.  
왜 딥러닝으로 바뀌었냐면 1940~2000 까지는 쓰였더 행렬이 하나 혹은 두 세개 밖에 되지 않았기에 딥하지 않았다. 하지만 요즘은 매우 많아졌다. 그 이유는 크게 두가지이다. 첫번쨰는 인터넷의 발달로 데이터가 엄청나게 많아졌다. 데이터가 많다는 것은 행렬 또한 많아져야 한다는 것을 의미. 즉 폭발적인 데이터량의 증가가 첫번째 이유이다. 딥한 것을 충분히 충족할 수 있는 하드웨어적인 발전도 있었다. 그것이 바로 그래픽카드이다. 원래는 그래픽카드를 우려한 rendering 때문에 만들엇다. 화면이 자연스럽고 부드럽게 움직이기 위해 많은 계산을 해야 한다. 모니터를 보면 1600*900 이렇게 픽셀 형태로 되어있다. 이 자체도 행렬이다. 현재 화면이 다음 화면으로 유려하게 바뀌게 하기 위한 계산이 행렬 계산이다. 필요한 계산은 *랑 +이다. 복잡한 계산이 아니다. 원래 cpu가 컴퓨토의 뇌이다. 하지만 cpu는 *,+ 하기에는 너무 머리가 좋다. core는 뇌가 몇개인지 알려준다. 매우 천재적인 뇌 8개한테 *,+를 시키기에는 너무 아깝다. 그래서 cpu가 아닌 좀 더 저렴한 gpu를 쓰기 시작했다. 초등학생 수준의 뇌이다.  
2000년대 초반에 계속 계산을 cpu로 하다가 gpu로 해볼까하는 발상의 전환이 생겼다. cpu로 1년 걸릴 것을 gpu로 했더니 1주일만에 하기 시작했다. 이때부터 ai시대가 도래했다.

딥러닝 그림  
동그라미와 화살표로 구성  
x -a-> y  y=ax  x는 입력값이고 a는 입력값과 곱해져 y라는 출력값이 나온다.  
y = ax + b를 하기 위해서는 입력값이 두개 필요하다. x라는 동그라미와 1이라는 동그라미를 준비한다. x에서 y로 향하는 화살표에는 a를 곱하고 1에서 y로 가는 화살표에는 b를 곱한다.   
y=ax+b가 ai의 최소형태이다.  
입력이 두 개일 때  
y= ax1 + bx2 + c  
입력이 세 개일 때  
y= ax1 + bx2 + cx3 + d  
입력 베타는 계속해서 늘릴 수 있다.  
만약 출력이 두개라면 하나의 입력 값에서 나가는 화살표가 두개. 화살표마다 할당된 숫자가 있기 때문에 입력값과 할당된 수를 곱하고 더하는 과정을 두 번 하면 된다.  
간단한 딥러닝 예제  
x1= 수면 시간 x2= 운동시간 x3= 칼로리 섭취량  
y1= 체중 y2= 혈압  
이 구조 자체를 모델이라고 부른다.  행렬=화살표=돈=ai=함수  
행렬을 구하기 위해서는 많은 데이터가 필요하다(빅데이터)  그러기 위해서 설문조사를 한다. 10000명의 사람들로부터 빅데이터를 받았고 이를 바탕으로 가운데 행렬을 잘 만들어야 하는데 이를 훈련이라고 한다.
