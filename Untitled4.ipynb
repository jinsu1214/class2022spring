```python
def test(CMPNM): 
  URL= ('http://apis.data.go.kr/6430000/cbRecreationalFoodInfoService/getRecreationalFoodInfo?CMPNM='+CMPNM+'&serviceKey=E4Emnq3vqP7veXrsgGXfs8CXHBVs6hADJSKFxOvjdpZFwk8Dja%2B2S1sd9tyFcGgQ3xsF2PeYw%2BAlLpC5ILk2Zg%3D%3D&currentPage=1&perPage=10&type=json')
  
  response = requests.get(URL)
  r = response.json()
  items = r['body']

  ADRES = [item['ADRES'] for item in items]
  CMPNM = [item['CMPNM'] for item in items]
  DC = [item['DC'] for item in items]
  LA = [item['LA'] for item in items]
  LO = [item['LO'] for item in items]
  MENU = [item['MENU'] for item in items]
  MNMNU = [item['MNMNU'] for item in items]
  SE = [item['SE'] for item in items]
  SIGUN = [item['SIGUN'] for item in items]
  TELNO = [item['TELNO'] for item in items]
  TIME = [item['TIME'] for item in items]
  _URL = [item['_URL'] for item in items]

  ADRES = pd.Series(ADRES)
  CMPNM = pd.Series(CMPNM)   
  DC = pd.Series(DC)
  LA = pd.Series(LA)
  LO = pd.Series(LO)
  MENU = pd.Series(MENU)
  MNMNU = pd.Series(MNMNU)
  SE = pd.Series(SE)
  SIGUN = pd.Series(SIGUN)
  TELNO = pd.Series(TELNO)
  TIME = pd.Series(TIME)
  _URL = pd.Series(_URL)

  df = pd.DataFrame({ 'ADRES': ADRES, 'CMPNM': CMPNM, 'DC': DC, 'LA': LA, 'LO': LO, 'MENU': MENU, 'MNMNU': MNMNU, 'SE': SE, 'SIGUN': SIGUN, 'TELNO': TELNO, 'TIME': TIME, '_URL': _URL})
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['지선생쌈촌', '옥천해뜨는집가든', '배바우손두부', '과수원가든', '옛장터 숯불갈비', '금옥식당', '보은누룽지백숙', '기사님식당', '신라식당']), outputs=['dataframe'])
iface.launch(debug=True)
```
